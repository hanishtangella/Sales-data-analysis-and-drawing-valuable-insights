```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import os
```

```python
all_month_data = pd.DataFrame()
```

```python
files = [file for file in os.listdir('./Sales_Data')]
for file in files:
    ds = pd.read_csv('./Sales_Data/' + file)
    all_month_data = pd.concat([all_month_data , ds])
    all_month_data.to_csv('all_data2.csv' , index= False)
```

```python
all_data = pd.read_csv('./all_data2.csv')
all_data
```

```python
all_data ['Month'] = all_data['Order Date'].str[0:2]
all_data
```

```python
na = all_data[all_data.isna().any(axis = 1)]
na
```

```python
all_data = all_data.dropna(how = 'all')
```

```python
all_data
```

```python
all_data = all_data[all_data['Month'] != 'Or']
all_data
```

```python
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype('int32')
all_data['Price Each'] = all_data['Price Each'].astype('float64')
```

```python
all_data['Total Price'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data
```

```python
results = all_data.groupby('Month').sum()
results
```

```python
months = range(1,13)
sales_data = results['Total Price']
plt.xticks(months)
plt.bar(months , sales_data)
plt.show()
```

# at which city maximum sales occured

```python
all_data
```

```python
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]
```

```python
all_data['City'] = all_data['Purchase Address'].apply(lambda x : get_city(x) + '(' + get_state(x) + ')')
all_data
```

```python
results = all_data.groupby('City').sum().sort_values('City')
results 
```

```python
cities = [city for city , df in all_data.groupby('City')]
plt.bar(cities,results['Total Price'])
plt.xticks(cities , rotation = 'vertical' , size = 10)
plt.xlabel('City')
plt.ylabel('Total sales (USD)')
plt.show()
```

```python
all_data
```

```python
all_data['Hour'] = pd.to_datetime(all_data['Order Date'])
all_data
```

```python
all_data = all_data.drop(columns = 'Hour')
all_data
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data
```

```python
results = all_data.groupby('Hour').count()['Quantity Ordered']
results
```

```python
hours = [hour for hour,df in all_data.groupby('Hour')]
plt.bar(hours,results)
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Quantity')
plt.show()
```
